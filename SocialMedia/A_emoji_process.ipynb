```python
import pandas as pd

data=pd.read_csv('data/Emoji æƒ…ç»ªåˆ†æåˆ—è¡¨ _ EmojiAll.csv')
data.columns=['emoji','Confidence','negative','neutral','positive']
data
```

```python
import pandas as pd

# ç¡®ä¿æ‰€æœ‰å€¼éƒ½æ˜¯å­—ç¬¦ä¸²ç±»å‹
data['emoji'] = data['emoji'].astype(str)
data['Confidence'] = data['Confidence'].astype(str)
data['negative'] = data['negative'].astype(str)
data['neutral'] = data['neutral'].astype(str)
data['positive'] = data['positive'].astype(str)

# åº”ç”¨ lambda å‡½æ•°
data['pure_emoji'] = data['emoji'].apply(lambda x: x[:1])
data['Confidence'] = data['Confidence'].apply(lambda x: x[5:])
# å¯¹P_valueæˆªå–Â±åŠåé¢çš„æ•°å­—
data['Confidence'] = data['Confidence'].apply(lambda x: x.split('Â±')[0])

# å¤„ç†æ— æ•ˆçš„å­—ç¬¦ä¸²
def safe_float_conversion(value):
    try:
        return float(value) / 100.0
    except ValueError:
        return None  # æˆ–è€…è¿”å›ä¸€ä¸ªé»˜è®¤å€¼ï¼Œå¦‚ 0.0

data['negative'] = data['negative'].apply(lambda x: x[5:-1]).apply(safe_float_conversion)
data['neutral'] = data['neutral'].apply(lambda x: x[5:-1]).apply(safe_float_conversion)
data['positive'] = data['positive'].apply(lambda x: x[5:-1]).apply(safe_float_conversion)

data
```

```python
data.dropna(subset=['pure_emoji'],inplace=True)
data.drop_duplicates(inplace=True)
data = data[data['pure_emoji'].notnull()]
data
```

```python
data.to_csv('data/emoji_data.csv', index=False,encoding='utf-8-sig')
```
