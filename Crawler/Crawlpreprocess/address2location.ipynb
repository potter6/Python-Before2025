```python
import requests
import pandas as pd
import os
import time
import transformer
from time import sleep
import random
import numpy as np

# 导入CSV
amp_api_data = pd.read_csv(r'poinameaddress.csv', encoding='utf-8-sig')

# 检查和处理没有上海市或相应地址的行
amp_api_data['address'] = amp_api_data['address'].apply(lambda x: '上海市' + str(x) if not str(x).startswith('上海市') else str(x))

# 请求地址前缀
req_url_pref = "https://restapi.amap.com/v3/geocode/geo?"
amp_api_key = '61c256acd1b853e34a6b9a31033c399d'
result = pd.DataFrame()  # 初始化
count = 0

def get_poi_from_amap(amp_api_name, amp_api_address):
    """
    从高德地图API下载地理编码
    """
    global result
    global count
    rep_params = {
        "key": amp_api_key,
        "address": amp_api_address
    }
    response = requests.get(req_url_pref, params=rep_params)
    data = response.json()  # 返回字典数据dict
    status = data["status"]
    
    if status != "0" and len(data["geocodes"]) > 0:
        geocode = data["geocodes"][0]
        lng = float(geocode["location"].split(",")[0])
        lat = float(geocode["location"].split(",")[1])
        wgs84_lng, wgs84_lat = transformer.gcj02towgs84(lng, lat)
        return {
            "name": amp_api_name,
            "address": amp_api_address,
            "outputaddress": geocode["formatted_address"],
            "gcj02_lng": lng,
            "gcj02_lat": lat,
            "wgs84_lng": wgs84_lng,
            "wgs84_lat": wgs84_lat
        }
    else:
        count += 1
        print(count, ":", amp_api_name, "未能在地图中找到位置")
        return {
            "name": amp_api_name,
            "address": amp_api_address,
            "outputaddress": np.nan,
            "gcj02_lng": np.nan,
            "gcj02_lat": np.nan,
            "wgs84_lng": np.nan,
            "wgs84_lat": np.nan
        }

# 处理缺失数据
for i in range(len(amp_api_data)):
    row = amp_api_data.iloc[i]
    if pd.isna(row['outputaddress']) or pd.isna(row['gcj02_lng']) or pd.isna(row['gcj02_lat']) or pd.isna(row['wgs84_lng']) or pd.isna(row['wgs84_lat']):
        amp_api_name = row['name']
        amp_api_address = row['address']
        new_data = get_poi_from_amap(amp_api_name, amp_api_address)
        for key, value in new_data.items():
            amp_api_data.at[i, key] = value
        sleep(random.randint(1, 3))

# 将结果写入到CSV中
output_path = os.getcwd() + os.sep + "location_gd_with_updates.csv"
amp_api_data.to_csv(output_path, index=False, encoding='utf-8-sig')
print("---------------经纬度获取完成(来源：高德地图)-------------------")
print("有", count, "个地址未查询到对应的经纬度")
```
