```python
import geopandas as gpd
import transbigdata as tbd

# 读取json文件
gdf = gpd.read_file('data_input\上海市.json')


gdf['geometry'] = gdf['geometry'].apply(lambda x: x.wkt)  # 将几何对象转换为WKT格式
gdf['geometry'] = gdf['geometry'].apply(lambda x: tbd.gcj02towgs84(x['lng'], x['lat']) if 'lng' in x and 'lat' in x else None)

# 重新输出json文件
with open('output_geojson_file.json', 'w', encoding='utf-8') as file:
    file.write(gdf.to_json())
```
