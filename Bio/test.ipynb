```python
import matplotlib.pyplot as plt
from scipy import interpolate
import numpy as np

import matplotlib.font_manager as mpt
zhfont=mpt.FontProperties(fname="C:\\Windows\\Fonts\\Microsoft YaHei UI") #

#
plt.grid(color='y',linestyle='--',linewidth=0.5)

#
# file = 'data.txt'
file="C:\\Users\\name\\Desktop\\.txt"
a = np.loadtxt(file)
# 
x = a[:,0]  # 
y = a[:,1]  # 
# 
tck = interpolate.splrep(x,y)
xx = np.linspace(min(x),max(x),100)
yy = interpolate.splev(xx,tck,der=0)
print(xx)
# 
plt.plot(x,y,'o',xx,yy)
plt.legend(['height','distance'])
plt.axis([-10,330,60,100])
plt.xlabel('Distance') #
plt.ylabel('Height')
plt.title('')
# 
plt.savefig('out.jpg')
plt.show()
```
