```python
import pandas as pd

# 假设data数据集以excel格式存储，读取数据集
data = pd.read_excel(r'D:\Code\Python\20241202合并-期刊论文爬虫\output\2006_21-23_result.xlsx')

# 假设source数据集以csv格式存储，读取数据集
source = pd.read_csv(r'D:\Code\Python\20241202合并-期刊论文爬虫\locationSource\archive\cities.csv')

# 提取地址列中的地名，并匹配source中的name列
def extract_and_match_city(address, source_df):
    # parts = address.split(';')
    parts = address
    city_names = []
    latitudes = []
    longitudes = []
    for part in parts:
        # 假设地名在地址的倒数第三部分
        city_name = part
        city_names.append(city_name)
        
        # 从source数据集中匹配地名
        match = source_df[source_df['name'] == city_name]
        if not match.empty:
            latitudes.append(match['latitude'].values[0])
            longitudes.append(match['longitude'].values[0])
        else:
            latitudes.append(None)
            longitudes.append(None)
    
    # 如果有多个地名，返回第一个匹配到的经纬度
    if latitudes and longitudes:
        return pd.Series([latitudes[0], longitudes[0]], index=['latitude', 'longitude'])
    else:
        return pd.Series([None, None], index=['latitude', 'longitude'])

# 应用函数查找精度和纬度
data[['latitude', 'longitude']] = data['Addresses'].apply(lambda x: extract_and_match_city(x, source))

# 删除多余的列
data.drop(columns=['Addresses'], inplace=True)

# 保存结果
data.to_csv('data_with_lat_lon.csv', index=False, encoding='utf-8-sig')
```
